Theory Questions 

1) What is hypothesis testing in statistics ?
>  Hypothesis testing in statistics is a formal procedure used to determine whether there is enough evidence in a sample of data to infer that a certain condition holds true for the entire population. It helps in making decisions about a population parameter based on sample data.

2) What is the null hypothesis, and how does it differ from the alternative hypothesis ?
>  The null hypothesis of a test always predicts no effect or no relationship between variables, while the alternative hypothesis states your research prediction of an effect or relationship.

3) What is the significance level in hypothesis testing, and why is it important ?
>  The significance level (alpha) in hypothesis testing is the threshold used to decide whether to reject the null hypothesis (H0). It represents the probability of making a Type I error, which occurs when the null hypothesis is true but is incorrectly rejected.\
> significance level importance:
a) Controls the Risk of Error: alpha limits the likelihood of rejecting H0 when it is actually true, ensuring the decision-making process is rigorous and reliable.
b)Sets the Standard of Evidence: A lower alpha(a) (e.g., 0.01) requires stronger evidence to reject H0, while a higher alpha(a) (e.g., 0.1) allows weaker evidence to suffice.
c)Balances Trade-offs: The choice of alpha(a) involves a trade-off between the risk of Type I error and Type II error (failing to reject H0 when it is false).

4) What does a P-value represent in hypothesis testing ?
>  In hypothesis testing, a p-value is a probability that measures how likely it is that the observed results occurred by chance. It's used to decide whether to reject the null hypothesis, which is the initial assumption about two groups in a population.

5) How do you interpret the P-value in hypothesis testing ?
>  A p-value less than (<) 0.05 is typically considered to be statistically significant, in which case the null hypothesis should be rejected. A p-value greater than (>) 0.05 means that deviation from the null hypothesis is not statistically significant, and the null hypothesis is Fail to rejecte.

6) What are Type 1 and Type 2 errors in hypothesis testing ?
>  In statistical hypothesis testing, a type I error, or a false positive, is the rejection of the null hypothesis when it is actually true. A type II error, or a false negative, is the failure to reject a null hypothesis that is actually false.

7) What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?
>  A one-tailed test in hypothesis testing only examines whether a parameter is significantly greater or less than a specified value (i.e., it looks for a change in one direction only), while a two-tailed test examines whether the parameter is significantly different from a specified value in either direction (greater or less than) essentially checking for any significant change without specifying a direction.

8) What is the Z-test, and when is it used in hypothesis testing ?
>  A z-test is a statistical test used to determine whether two population means are different when the variances are known and the sample size is large. It can also be used to compare one mean to a hypothesized value. The data must approximately fit a normal distribution, otherwise the test doesn't work.

9) How do you calculate the Z-score, and what does it represent in hypothesis testing ?
>  A Z-score is calculated by subtracting the population mean from a data point and then dividing the result by the population standard deviation, in hypothesis testing, a Z-score represents how many standard deviations a data point is away from the population mean.

10) What is the T-distribution, and when should it be used instead of the normal distribution ?
>   A T-distribution, also called "Student's t-distribution", is a statistical distribution similar to the normal distribution but with heavier tails, making it more suitable for situations where the sample size is small and the population standard deviation is unknown.

11) What is the difference between a Z-test and a T-test ?
>   z-tests are used when we have large sample sizes (n > 30), whereas t-tests are most helpful with a smaller sample size (n < 30). Both methods assume a normal distribution of the data, but the z-tests are most useful when the standard deviation is known.

12) What is the T-test, and how is it used in hypothesis testing ?
>   A T-test is a statistical test used in hypothesis testing to compare the means of two groups (samples), determining if there is a statistically significant difference between them, typically by comparing the sample means to a null hypothesis that states there is no difference between the populations they represent.

13) What is the relationship between Z-test and T-test in hypothesis testing ?
>    The t-test is typically used when the sample size is small, generally less than 30. It is designed to be robust when the sample size does not meet the threshold needed for applying the Central Limit Theorem. The Z-test is used when the sample size is large, typically greater than 30.

14) What is a confidence interval, and how is it used to interpret statistical results ?
>   A confidence interval shows the probability that a parameter will fall between a pair of values around the mean. Confidence intervals show the degree of uncertainty or certainty in a sampling method. They are constructed using confidence levels of 95% or 99%.

15) What is the margin of error, and how does it affect the confidence interval ?
>   A "margin of error" is a statistical measure that represents the degree of uncertainty around a survey result or estimate, essentially indicating how much the result could vary from the true population value, while a "confidence interval" is the range of values within which the true population parameter is likely to fall, with the margin of error defining the "plus or minus" range around the estimated value within that confidence interval, meaning, a larger margin of error results in a wider confidence interval, signifying less precision in the estimate. 

16) How is Bayes' Theorem used in statistics, and what is its significance ?
>   Bayes' Theorem thus gives the probability of an event based on new information that is or may be related to that event. The formula also can be used to determine how the probability of an event occurring may be affected by hypothetical new information. It is significant because it enables more accurate predictions and inferences in situations where uncertainty exists, particularly when dealing with complex systems or limited data.

17) What is the Chi-square distribution, and when is it used ?
>   A Chi-square distribution is a statistical distribution used primarily for hypothesis testing, particularly when analyzing categorical data, to determine if there is a significant difference between observed frequencies and expected frequencies in a dataset, often used in tests like the "chi-square goodness of fit test" and the "chi-square test of independence" to assess relationships between categorical variables; its shape is determined by the "degrees of freedom" parameter which depends on the number of categories in the data being analyzed.

18) What is the Chi-square goodness of fit test, and how is it applied ?
>   A Chi-square goodness of fit test is a statistical test used to determine whether a set of observed data from a sample closely aligns with a theoretical distribution (like a normal distribution), by comparing the observed frequencies in different categories to the expected frequencies based on that theoretical distribution; essentially, it assesses how well the data "fits" the expected pattern; it is applied by calculating a Chi-square statistic based on the differences between observed and expected frequencies, and then comparing this value to a critical value to determine statistical significance.

19) What is the F-distribution, and when is it used in hypothesis testing ?
>   F test is a statistical test that is used in hypothesis testing to check whether the variances of two populations or two samples are equal or not. In an f test, the data follows an f distribution. This test uses the f statistic to compare two variances by dividing them.

20) What is an ANOVA test, and what are its assumptions ?
>   An ANOVA (Analysis of Variance) test is a statistical method used to compare the means of multiple groups, assuming that the data is normally distributed, samples are randomly selected and are independent, and all groups have roughly equal variances (homogeneity of variance), essentially, it checks if there is a significant difference between the means of different groups based on the variation within each group. 

21) What are the different types of ANOVA tests
>   ANOVA (Analysis of Variance) is a statistical method used to compare means across multiple groups. Different types of ANOVA tests exist based on factors like the number of independent variables and whether the samples are related or not. Here are the main types:
a) One-Way ANOVA : Compares the means of three or more independent (unrelated) groups based on a single independent variable.
b) Two-Way ANOVA : Examines the effect of two independent variables on a dependent variable, considering their interaction.
c) Factorial ANOVA : A generalization of two-way ANOVA for more than two independent variables. 

22) What is the F-test, and how does it relate to hypothesis testing ?
>   An F-test is a statistical test used in hypothesis testing to compare the variances of two populations or samples, essentially determining if there is a significant difference between the variability of two data sets by calculating the ratio of their variances.

```python
#1) Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis (H0): The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference detected.")

sample_data = [75, 80, 85, 90, 88, 76, 92, 95, 85, 78]  # Sample data
population_mean = 82  # Known population mean
population_std = 5  # Known population standard deviation

z_test(sample_data, population_mean, population_std)
```

```python
#4) Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def two_tailed_z_test(sample, population_mean, population_std, alpha=0.05):
    
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis (H0): The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference detected.")

    x = np.linspace(-4, 4, 1000)  # Z-score range
    y = stats.norm.pdf(x, 0, 1)  # Standard normal distribution

    critical_value = stats.norm.ppf(1 - alpha / 2)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label="Standard Normal Distribution", color='blue')

    plt.fill_between(x, y, where=(x <= -critical_value) | (x >= critical_value), color='red', alpha=0.5, label="Rejection Region")

    plt.axvline(z_score, color='green', linestyle='--', linewidth=2, label=f'Observed Z-score: {z_score:.2f}')
    
    plt.axvline(-critical_value, color='black', linestyle='dotted', linewidth=2, label=f'Critical Z: {critical_value:.2f}')
    plt.axvline(critical_value, color='black', linestyle='dotted', linewidth=2)

    plt.title("Two-Tailed Z-Test: Decision Regions")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

np.random.seed(42)
sample_data = np.random.normal(loc=82, scale=5, size=30)

population_mean = 80
population_std = 5

two_tailed_z_test(sample_data, population_mean, population_std)
```

```python
#5) Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_errors(alpha=0.05, effect_size=0.5, sample_size=30, population_std=5):

    standard_error = population_std / np.sqrt(sample_size)

    x = np.linspace(-4, 4, 1000)
    h0_distribution = stats.norm.pdf(x, 0, 1)

    h1_mean = effect_size
    h1_distribution = stats.norm.pdf(x, h1_mean, 1)

    critical_value = stats.norm.ppf(1 - alpha / 2)

    plt.fill_between(x, h0_distribution, where=(x >= critical_value) | (x <= -critical_value), 
                     color='red', alpha=0.5, label="Type 1 Error ()")

    beta_region = (x < critical_value) & (x > -critical_value)
    plt.fill_between(x, h1_distribution, where=beta_region, 
                     color='blue', alpha=0.5, label="Type 2 Error ()")

    plt.plot(x, h0_distribution, label="Null Hypothesis (H0)", color='black')
    plt.plot(x, h1_distribution, label="Alternative Hypothesis (H1)", color='green')

    plt.axvline(-critical_value, color='black', linestyle='dotted', linewidth=2, label=f'Critical Z: {critical_value:.2f}')
    plt.axvline(critical_value, color='black', linestyle='dotted', linewidth=2)

    plt.title("Type 1 & Type 2 Errors in Hypothesis Testing")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid()
    plt.show()

visualize_errors()
```

```python
#6) Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
import scipy.stats as stats

def independent_t_test(sample1, sample2, alpha=0.05):
    
    mean1, mean2 = np.mean(sample1), np.mean(sample2)
    std1, std2 = np.std(sample1, ddof=1), np.std(sample2, ddof=1)
    n1, n2 = len(sample1), len(sample2)

    t_statistic, p_value = stats.ttest_ind(sample1, sample2, equal_var=False)  # Welchs T-test

    print(f"Sample 1: Mean = {mean1:.4f}, Std Dev = {std1:.4f}, Size = {n1}")
    print(f"Sample 2: Mean = {mean2:.4f}, Std Dev = {std2:.4f}, Size = {n2}")
    print(f"T-score: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis (H0): The two sample means are significantly different.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference detected between the means.")

np.random.seed(42)
sample1 = np.random.normal(loc=75, scale=10, size=30)
sample2 = np.random.normal(loc=80, scale=10, size=30)

independent_t_test(sample1, sample2)
```

```python
#7) Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def paired_t_test(sample1, sample2, alpha=0.05):
    
    t_statistic, p_value = stats.ttest_rel(sample1, sample2)
    
    mean_diff = np.mean(sample2 - sample1)
    
    print(f"Mean Difference: {mean_diff:.4f}")
    print(f"T-score: {t_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis (H0): The mean difference between the pairs is significantly different.")
    else:
        print("Fail to reject the null hypothesis (H0): No significant difference in the mean between the pairs.")
    
    differences = sample2 - sample1
    plt.figure(figsize=(8, 6))
    plt.bar(range(len(differences)), differences, color='purple', alpha=0.7)
    plt.axhline(0, color='black', linewidth=1, linestyle='--')
    plt.xlabel('Paired Samples')
    plt.ylabel('Difference (Post - Pre)')
    plt.title('Paired Sample T-Test: Difference Between Paired Samples')
    plt.grid(axis='y')
    plt.show()

np.random.seed(42)
sample1 = np.random.normal(loc=80, scale=5, size=30)
sample2 = sample1 + np.random.normal(loc=5, scale=3, size=30)

paired_t_test(sample1, sample2)
```

```python
#9) Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence_level=0.95):
    
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)

    if sample_size > 30:
        critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2) 
    else:
        critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df=sample_size - 1) 
    
    standard_error = sample_std / np.sqrt(sample_size)
    
    margin_of_error = critical_value * standard_error
    ci_lower = sample_mean - margin_of_error
    ci_upper = sample_mean + margin_of_error
    
    return ci_lower, ci_upper

np.random.seed(42) 
sample_data = np.random.normal(loc=100, scale=15, size=50) 

ci_lower, ci_upper = confidence_interval(sample_data, confidence_level=0.95)
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#10) Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
import scipy.stats as stats

def margin_of_error(sample, confidence_level=0.95):
    
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)

    if sample_size > 30: 
        critical_value = stats.norm.ppf(1 - (1 - confidence_level) / 2)
    else:  
        critical_value = stats.t.ppf(1 - (1 - confidence_level) / 2, df=sample_size - 1)
    
    standard_error = sample_std / np.sqrt(sample_size)
    
    margin_error = critical_value * standard_error
    
    return margin_error

np.random.seed(42) 
sample_data = np.random.normal(loc=100, scale=15, size=50)  

moe = margin_of_error(sample_data, confidence_level=0.95)
print(f"Margin of Error: {moe:.2f}")
```

```python
#11) Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior, likelihood, evidence):
    
    posterior = (likelihood * prior) / evidence
    return posterior

# P(H): Prior probability of having the disease (e.g., 1 in 1000 people have the disease)
prior = 1 / 1000  

# P(E|H): Likelihood of testing positive given the person has the disease
likelihood = 0.99  # 99% chance of testing positive if the person has the disease

# P(E): Probability of a positive test result (sensitivity + false positives)
evidence = 0.05  # 5% of the population tests positive (including both true positives and false positives)

posterior = bayes_theorem(prior, likelihood, evidence)

print(f"Posterior Probability (P(H|E)): {posterior:.5f}")
```

```python
#13) Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

import numpy as np

def calculate_expected_frequencies(observed_data):
    
    observed = np.array(observed_data)
    
    # Calculate the row sums and column sums
    row_sums = observed.sum(axis=1)  # Sum of each row
    col_sums = observed.sum(axis=0)  # Sum of each column
    
    total = observed.sum()
    
    expected = np.outer(row_sums, col_sums) / total
    
    return expected

observed_data = [
    [30, 10],  # Male: Product A, Product B
    [20, 40]   # Female: Product A, Product B
]


expected_frequencies = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:\n", np.array(observed_data))
print("\nExpected Frequencies:\n", expected_frequencies)
```

```python
#14) Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
import scipy.stats as stats

def goodness_of_fit_test(observed, expected):

    # Calculate the Chi-square statistic and p-value
    chi2_stat, p_value = stats.chisquare(observed, expected)
    return chi2_stat, p_value

observed = [15, 18, 20, 25, 12, 10]
expected = [sum(observed) / 6] * 6

# goodness-of-fit test
chi2_stat, p_value = goodness_of_fit_test(observed, expected)

print(f"Chi-square statistic: {chi2_stat}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The observed data does not follow the expected distribution.")
else:
    print("\nFail to reject the null hypothesis: The observed data follows the expected distribution.")
```

```python
#15) Create a Python script to simulate and visualize the Chi-square distribution and discuss its characteristics.

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Function to simulate and plot the Chi-square distribution
def plot_chi_square_distribution(df_values, num_samples=1000, bins=30):

    plt.figure(figsize=(10, 6))
    
    for df in df_values:
        chi_square_data = np.random.chisquare(df, num_samples)
        
        plt.hist(chi_square_data, bins=bins, density=True, alpha=0.6, label=f'df={df}')
    
        x = np.linspace(0, np.max(chi_square_data), 100)
        plt.plot(x, stats.chi2.pdf(x, df), color='black', lw=2)
    
    plt.title("Chi-square Distribution for Various Degrees of Freedom")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

df_values = [1, 5, 10]
plot_chi_square_distribution(df_values)
```

```python
#16) Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
import scipy.stats as stats

def f_test_variance_comparison(sample1, sample2):
    
    # Calculate the variances of both samples
    var1 = np.var(sample1, ddof=1) 
    var2 = np.var(sample2, ddof=1)
    
    # F-statistic (larger variance / smaller variance)
    f_statistic = var1 / var2 if var1 > var2 else var2 / var1
    
    # Perform the F-test (two-tailed)
    df1 = len(sample1) - 1 
    df2 = len(sample2) - 1 
    p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))
    
    return f_statistic, p_value

np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=100)
sample2 = np.random.normal(loc=50, scale=20, size=100)

f_statistic, p_value = f_test_variance_comparison(sample1, sample2)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: The variances are significantly different.")
else:
    print("\nFail to reject the null hypothesis: The variances are not significantly different.")
```

```python
#17) Write a Python program to perform an ANOVA test to compare means between multiple groups and interpret the results.

import numpy as np
import scipy.stats as stats

def perform_anova(*groups):
    f_statistic, p_value = stats.f_oneway(*groups) # Perform one-way ANOVA
    return f_statistic, p_value

np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)
group3 = np.random.normal(loc=60, scale=10, size=30)

f_statistic, p_value = perform_anova(group1, group2, group3)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant difference between group means.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between group means.")
```

```python
#18) Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def perform_anova(*groups):
    
    f_statistic, p_value = stats.f_oneway(*groups) # Perform one-way ANOVA
    return f_statistic, p_value

np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)  
group2 = np.random.normal(loc=55, scale=10, size=30)  
group3 = np.random.normal(loc=60, scale=10, size=30)  
group4 = np.random.normal(loc=65, scale=10, size=30)  

f_statistic, p_value = perform_anova(group1, group2, group3, group4)

print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

alpha = 0.05
if p_value < alpha:
    print("\nReject the null hypothesis: There is a significant difference between group means.")
else:
    print("\nFail to reject the null hypothesis: There is no significant difference between group means.")

groups = [group1, group2, group3, group4]
group_labels = ['Group 1', 'Group 2', 'Group 3', 'Group 4']

# Plotting
plt.figure(figsize=(10, 6))

# Boxplot
plt.subplot(1, 2, 1)
sns.boxplot(data=groups, palette="Set3")
plt.title('Boxplot of Group Distributions')
plt.xticks(ticks=np.arange(4), labels=group_labels)

# Violin plot
plt.subplot(1, 2, 2)
sns.violinplot(data=groups, palette="Set3")
plt.title('Violin Plot of Group Distributions')
plt.xticks(ticks=np.arange(4), labels=group_labels)

plt.tight_layout()
plt.show()
```

```python
#19) Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def check_anova_assumptions(*groups):
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)
        print(f"Group {i+1} normality test (Shapiro-Wilk): stat={stat:.4f}, p-value={p_value:.4f}")
        if p_value < 0.05:
            print(f"  - Group {i+1}: Data is NOT normal.")
        else:
            print(f"  - Group {i+1}: Data is normal.")
    
    # Check for equal variance using Levene's test
    stat, p_value = stats.levene(*groups)
    print(f"\nLevene's test for equal variance: stat={stat:.4f}, p-value={p_value:.4f}")
    if p_value < 0.05:
        print("  - Variances are NOT equal.")
    else:
        print("  - Variances are equal.")
    
group1 = np.random.normal(loc=10, scale=5, size=50)
group2 = np.random.normal(loc=10, scale=5, size=50)
group3 = np.random.normal(loc=10, scale=5, size=50)

check_anova_assumptions(group1, group2, group3)
```

```python
#20) Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

np.random.seed(42)

factor_A = np.random.choice(['A1', 'A2'], size=100)
factor_B = np.random.choice(['B1', 'B2'], size=100)

# Dependent variable (Continuous)
dependent_var = np.random.normal(50, 10, 100) + 10 * (factor_A == 'A2') + 5 * (factor_B == 'B2')

data = pd.DataFrame({'Factor_A': factor_A, 'Factor_B': factor_B, 'Dependent_Variable': dependent_var})

# Perform Two-Way ANOVA
model = ols('Dependent_Variable ~ C(Factor_A) + C(Factor_B) + C(Factor_A):C(Factor_B)', data=data).fit()
anova_results = anova_lm(model)

sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

sns.boxplot(x='Factor_A', y='Dependent_Variable', hue='Factor_B', data=data, palette="Set2")
plt.title('Two-Way ANOVA: Interaction between Factor A and Factor B')
plt.show()
```

```python
#21)  Write a Python program to visualize the F-distribution and discuss its use in hypothesis testing.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def plot_f_distribution(df1, df2):
    x = np.linspace(0, 5, 1000)
    y = stats.f.pdf(x, df1, df2)
    
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label=f'F-distribution (df1={df1}, df2={df2})', color='blue')
    plt.fill_between(x, y, alpha=0.2, color='blue')
    plt.axvline(stats.f.ppf(0.95, df1, df2), color='red', linestyle='--', label='Critical value (=0.05)')
    
    plt.title(f'F-Distribution with df1={df1} and df2={df2}')
    plt.xlabel('F-value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid()
    plt.show()

plot_f_distribution(5, 10)
```

```python
#22) Perform a one-way ANOVA test in Python and visualize the results with boxplots to compare group means.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def one_way_anova_test():
    np.random.seed(42)  # Generate sample data for three groups
    group1 = np.random.normal(50, 10, 30)
    group2 = np.random.normal(55, 10, 30)
    group3 = np.random.normal(60, 10, 30)
    
    f_stat, p_value = stats.f_oneway(group1, group2, group3)
    print(f'F-statistic: {f_stat:.4f}, p-value: {p_value:.4f}')
    
    data = [group1, group2, group3]
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=data)
    plt.xticks([0, 1, 2], ['Group 1', 'Group 2', 'Group 3'])
    plt.title('One-Way ANOVA Test: Comparison of Group Means')
    plt.ylabel('Values')
    plt.grid()
    plt.show()

one_way_anova_test()
```

```python
#24)  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

import numpy as np
import scipy.stats as stats

def variance_hypothesis_test():
    np.random.seed(42)
    sample = np.random.normal(50, 10, 30)
    
    sigma_squared = 100
    
    sample_var = np.var(sample, ddof=1)
    chi_square_stat = (len(sample) - 1) * sample_var / sigma_squared
    
    p_value = 1 - stats.chi2.cdf(chi_square_stat, df=len(sample) - 1)
    
    if p_value < 0.05:
        print("Reject nnull hypothesis: The sample variance is significantly different from the hypothesized variance.")
    else:
        print("Fail to reject null hypothesis: No significant difference in variance.")

variance_hypothesis_test()
```

```python
#25) Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.

import numpy as np
import statsmodels.api as sm

# Sample data
successes_1 = 50 # Group 1: Successes and total observations
total_1 = 100

successes_2 = 30 # Group 2: Successes and total observations
total_2 = 100

# Proportions
p1 = successes_1 / total_1
p2 = successes_2 / total_2

# Perform Z-test for proportions
count = np.array([successes_1, successes_2])
nobs = np.array([total_1, total_2])

z_stat, p_value = sm.stats.proportions_ztest(count, nobs)

print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

alpha = 0.05 
if p_value < alpha:
    print("Reject the null hypothesis: The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the proportions.")
```

```python
#26) Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

# Sample data for two groups
data1 = np.random.normal(loc=10, scale=5, size=100)  # Group 1 with variance 25 (5^2)
data2 = np.random.normal(loc=10, scale=3, size=100)  # Group 2 with variance 9 (3^2)

# Perform the F-test (variance ratio test)
var1 = np.var(data1, ddof=1)  # Sample variance of data1
var2 = np.var(data2, ddof=1)  # Sample variance of data2

# Calculate the F-statistic (larger variance / smaller variance)
f_statistic = var1 / var2 if var1 > var2 else var2 / var1
df1 = len(data1) - 1  # Degrees of freedom for data1
df2 = len(data2) - 1  # Degrees of freedom for data2

# p-value from the F-distribution
p_value = 1 - stats.f.cdf(f_statistic, df1, df2)

# Output the results
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in the variances.")

# Visualize the data distributions using boxplots
plt.figure(figsize=(8, 6))
plt.boxplot([data1, data2], labels=['Group 1', 'Group 2'])
plt.title('Boxplot Comparison of Two Groups')
plt.ylabel('Values')
plt.show()
```

```python
#27) Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

np.random.seed(42)  # For reproducibility
observed_data = np.random.randint(1, 7, size=1000)  # Simulated 1000 die rolls

# Calculate the observed frequencies
observed_freq = [np.sum(observed_data == i) for i in range(1, 7)]

# The expected frequencies for a fair die (1000 rolls, so each outcome should have a frequency of 1000/6)
expected_freq = [1000 / 6] * 6

chi2_stat, p_value = stats.chisquare(observed_freq, expected_freq)

alpha = 0.05 
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")

labels = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(labels))

plt.figure(figsize=(8, 6))
plt.bar(x - 0.2, observed_freq, 0.4, label='Observed', color='b')
plt.bar(x + 0.2, expected_freq, 0.4, label='Expected', color='r', alpha=0.5)

plt.xticks(x, labels)
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies for Die Rolls')
plt.legend()
plt.show()
```
